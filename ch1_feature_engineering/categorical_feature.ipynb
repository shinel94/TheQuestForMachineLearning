# 범주형 피처
> 범주형 피처는 입력 독립변수의 도메인이 수치형 처럼 일정 무한한 샘플을 가진 구간이 아니라 <br>
> 선택할 수 있는 값이 한정된 유한한 선택범위를 가지는 독립변수를 의미합니다.<br>
> 보통 문자열이나 level또는 categori와 같은 형식으로 전달이 되는데, 소수의 기계학습 알고리즘(트리 기반 모델)에서만 이런 범주형 피쳐를 그대로 모델 입력으로 사용할 수 있습니다 <br>
> 다른 기계학습 알고리즘(로지스틱 회귀 등)을 활용하기 위해서는 이런 범주형 피쳐를 반드시 수치형 피쳐로 변경을 한뒤 학습을 진행해야 합니다. <br>
> 이때 수치형 피쳐로 변형하는 방식에는 순서형 인코딩 / 원핫인코딩 / 이진 인코딩 등이 있습니다.
>> 순서형 인코딩 : 카테고리별로 순서대로 0~n까지의 값을 가지도록 할당합니다. 이때 1과 2 사이의 관계와 0과 n사이의 관계가 수치적으로 다름을 인지한 상태로 변환을 진행해야 합니다. <br>
>> 원핫 인코딩 : 순서형 인코딩에서 발생하는 데이터의 의도하지 않은 관계를 제거할 수 있는 방법으로 각 카테고리별로 해당 카테고리에 해당하는지 여부를 1과 0으로 표시하는 변환입니다. <br>
>> 이진 인코딩 : 이진 인코딩은 순서형 인코딩과 유사하게 각 값을 0~n사이로 할당한뒤 해당 값을 2진수 표현식으로 변경하여 해당 2진수 표현식을 특징 벡터로 변환합니다. 이때 원 핫 인코딩보다 적은 수의 메모리 공간을 필요하게 되면서 순서형 인코딩의 의도하지 않은 연관성을 줄일 수 있습니다.
